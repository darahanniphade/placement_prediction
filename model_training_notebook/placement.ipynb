# make all fields in numeric


```python
yes_no_map={'Yes':1,'No':0}
df['ExtracurricularActivities']= df['ExtracurricularActivities'].map(yes_no_map)

training_maper={'Yes':1,'No':0}
df['PlacementTraining']=df['PlacementTraining'].map(training_maper)

status= {'NotPlaced':0,"Placed":1}
df['PlacementStatus']=df['PlacementStatus'].map(status)

```

```python
df.tail()
```

```python
df=df.drop(columns=['StudentID'])
```

```python
df.head()
```

# select the input and the output feature

```python
x=df.iloc[:,0:10]
y=df.iloc[:,-1]
```

#  split the dataset


```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

# normalize the dataset

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

# import the model (as per your choice)

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf =LogisticRegression()
```

# train the model

```python
clf.fit(x_train,y_train)
```

# predict the results

```python
y_pred=clf.predict(x_test)
```

```python
y_test
```

#  use the accuracy function

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
